### **Load the Quiz Files**

Load quiz grades (quiz_1_grades.csv, ..., quiz_5_grades.csv)

```python
file_names = ["quiz_1_grades (1) - quiz_1_grades (1).csv.xls","quiz_2_grades - quiz_2_grades.csv.xls","quiz_3_grades - quiz_3_grades.csv.xls",
             "quiz_4_grades - quiz_4_grades.csv.xls","quiz_5_grades - quiz_5_grades.csv.xls"]
```

```python
#load all the quizzes now, but i want it the dataframe to be indexed on emails, and return all quizzes corresponding to that email.
#for the expected output of this code, refer to the submissional pdf shared with you in whatsapp group.
all_quiz_grades = pd.DataFrame()

# Loop through each specified CSV file
for file in file_names:
    # Read the CSV file
    df = pd.read_csv(file)
    
    # Standardize the email addresses to lowercase
    df['Email'] = df['Email'].str.lower()
    
    # Select only necessary columns (assuming columns include 'Email Address' and 'Grade')
    df = df[['Email', 'Grade']]
    
    # Extract quiz number from the file name
    quiz_number = file.split('_')[1]  # e.g., 'quiz_1_grades.csv' -> '1'
    
    # Rename the 'Grade' column to include the quiz number
    df.rename(columns={'Grade': f'Quiz_{quiz_number}'}, inplace=True)
    
    # Set 'Email Address' as index
    df.set_index('Email', inplace=True)
    
    # Merge the current quiz DataFrame with the all_quiz_grades DataFrame
    if all_quiz_grades.empty:
        all_quiz_grades = df
    else:
        all_quiz_grades = all_quiz_grades.join(df, how='outer')

# Display the combined DataFrame
all_quiz_grades
```

### Homework Grades

**Question 1: Identifying Homework Columns**

```python
#Identify and separate the columns representing homework scores and their respective maximum points.
columns=list(filter(lambda col: 'Homework' in col, final_df.columns))
max_homework = list(filter(lambda col: 'Max' in col, final_df[columns].columns))
homework = set(columns)-set(max_homework)
homework= list(homework)

```

**Question 2: Calculating Homework Score by Total Method**

```python
#Calculate the homework scores by summing the actual scores and dividing by the total possible points.
sum_homework = final_df[homework].sum(axis=1)
sum_max_homework = final_df[max_homework].sum(axis=1)
hw_by_total = sum_homework/sum_max_homework
```

**Question 3: Storing Homework Score by Total Method**

```python
#Add the calculated total homework scores to the DataFrame.
final_df['HW by Total'] = hw_by_total
```

**Question 4: Preparing Data for Average Calculation**

```python
#Prepare the maximum points data to align with the actual scores for element-wise division.
hw_max_data = final_df[max_homework]
hw_cols = final_df[homework]
hw_max_data = hw_max_data.set_axis(hw_cols.columns, axis=1)
```

**Question 5: Calculating Homework Score by Average Method**

```python
#Calculate the average homework scores by dividing each actual score by its corresponding maximum points, summing these ratios, and then dividing by the number of assignments.
homework_scores_ratio = hw_cols / hw_max_data
all_homework = homework_scores_ratio.sum(1)/len(homework)
```

**Question 6: Storing Homework Score by Average Method**

```python
#Add the calculated average homework scores to the DataFrame.
final_df['HW by  Average'] = all_homework
```

**Question 7: Determining the Final Homework Score**

```python
#Determine the final homework score for each student by taking the maximum value between the total and average methods.
final_df['Homework Score'] = final_df[['HW by Total','HW by  Average']].max(axis=1)
final_df
```

## Group the data to calculate final scores!

```python
weights = pd.Series(
    {
        "Exam 1 Score": 0.05,
        "Exam 2 Score": 0.10,
        "Exam 3 Score": 0.15,
        "Quiz Score": 0.30,
        "Homework Score": 0.40,
    }
)
```

```python
weights
```

1. Define the weights for each component.
2. Calculate the final score by applying the weights to each component and summing the results.
3. Calculate the ceiling score by rounding up the final score multiplied by 100.
4. Define a function to convert numeric scores to letter grades based on standard thresholds.
5. Map the ceiling scores to letter grades.
6. Store the letter grades as a categorical column in the DataFrame.
7. Display the final letter grades.

```python
#For Expected output, refer to the submissional pdf.
scores = final_df[['Exam 1 Score','Exam 2 Score','Exam 3 Score','Quiz Score','Homework Score']]
weighted_scores = scores * weights.values
weighted_scores = weighted_scores.sum(1)
ceil_score = np.ceil(weighted_scores*100)

def score_to_grade(score):
    if score >= 90:
        return 'A'
    elif score >= 80:
        return 'B'
    elif score >= 70:
        return 'C'
    elif score >= 60:
        return 'D'
    else:
        return 'F'

final_df['Grades'] = ceil_score.apply(score_to_grade)
final_df.set_index('SID',inplace=True)
final_df['Grades']
```
